```python
def read_gene_positions(file_path):
    genes = {}
    with open(file_path, "r") as file:
        for line in file:
            data = line.strip().split()
            chrom, start, stop, name = data[:4]
            if chrom not in genes:
                genes[chrom] = {}
            genes[chrom][name] = (int(start), int(stop))
    return genes
```

```python
def read_bed(filename):
    """lit un ficheir bed et retourne un dictionnaire avec les positions des gnes"""
    dico = {}
    with open(filename, "r") as file:
        for line in file:
            data = line.strip().split()
            chrom, start, stop, name = data[:4]
            if chrom not in genes:
                dico[chrom] = {}
            dico[chrom][name] = (int(start), int(stop))
    return dico
```
